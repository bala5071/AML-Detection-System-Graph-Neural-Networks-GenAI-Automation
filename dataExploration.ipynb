```python
import pandas as pd
import numpy as np

df = pd.read_csv('Data/HI-Small_Trans.csv')
print(df.head())
print(df.isnull().sum())
```

```python
print(df.info())
print(df.describe())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Receiving Currency'] = le.fit_transform(df['Receiving Currency'].astype(str))
df['Payment Currency'] = le.fit_transform(df['Payment Currency'].astype(str))
df['Payment Format'] = le.fit_transform(df['Payment Format'].astype(str))
print(df)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
print(df)
print(df['Timestamp'].min())
print(df['Timestamp'].max())
```

```python
df['Timestamp'] = df['Timestamp'].apply(lambda x: x.value)
df['Timestamp'] = (df['Timestamp']-df['Timestamp'].min())/(df['Timestamp'].max()-df['Timestamp'].min())
print(df)
```

```python
label_counts = df["Is Laundering"].value_counts()
label_counts

label_counts / label_counts.sum()
```

```python
df['Account'] = df['From Bank'].astype(str) + '_' + df['Account']
df['Account.1'] = df['To Bank'].astype(str) + '_' + df['Account.1']
df = df.sort_values(by=['Account'])
rf = df[['Account.1', 'Amount Received', 'Receiving Currency']]
pf = df[['Account', 'Amount Paid', 'Payment Currency']]
rf = rf.rename({'Account.1':'Account'}, axis=1)
currency = sorted(df['Receiving Currency'].unique())
print(df, rf, pf, currency)
```

```python
sdf = df[['Account', 'From Bank']]
rdf = df[['Account.1', 'To Bank']]
sdf = sdf.rename({'From Bank':'Bank'}, axis=1)
rdf = rdf.rename({'Account.1':'Account', 'To Bank':'Bank'}, axis=1)
s = df[df['Is Laundering'] == 1]
s1 = s[['Account', 'Is Laundering']]
s2 = s[['Account.1', 'Is Laundering']]
s2 = s2.rename({'Account.1':'Account'}, axis=1)
s = pd.concat([s1, s2], join='outer')
s = s.drop_duplicates()
df = pd.concat([sdf, rdf], join='outer')
df = df.drop_duplicates()
df['Is Laundering'] = 0
df.set_index('Account', inplace=True)
df.update(s.set_index('Account'))
df = df.reset_index()
print(df)
```
